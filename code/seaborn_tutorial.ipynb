# Load the libraries

Make sure you got `pandas`, `numpy` and `seaborn` installed. You should have these already installed if you're using anaconda.

- `re`: handling of regular expressions (comes by default with python)
- `numpy`: Manipulate numeric arrays and number cruching
- `pandas`: Flexible data wrangling
- `seaborn`: Nice data visualization

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="ticks")
```

# Load the data

We will examine the file `data/college_data.csv`. The file contains more than 160 different variables for the 395 doctoral universities in the US. The data was obtained from [Integrated Postsecondary Education Data System (IPEDS)](https://nces.ed.gov/ipeds/use-the-data). The variables correspond the most recent available academic year. In most of the cases, this mean 2018-19. However, some variables are from the 2017-18 year.

In general, `pandas` can read/write CSV, TSV, XSLX, ODS, and more. You might need extra libraries when dealing with spreadsheets. Check `openpyxl`.

`.head()` and `.tail()` help us display just the first/last results of our data set, to get an idea how the whole file looks like.

Data wrangling performed as in the previous pandas tutorial.

```python
data = pd.read_csv('../data/college_data.csv')
data.set_index('Institution Name', inplace=True)
data.drop(columns=['UnitID'], inplace=True)
data.rename(columns = lambda x : '_'.join(re.split("\s+", str(x), flags=re.UNICODE)[:-1]).translate(str.maketrans("'-/+!@#$%^&*()+=?",'_________________', "'-/&")), 
            inplace=True)
data.head()
```

# Statistical relationships

- Relating variables with scatterplots and lineplots
- The current example dataset has no timeseries data, so we'll skip lineplots.
- Object returned is a `FacetGrid`
- **Extremely painful** to edit the legend

```python
sns.set_context("poster")

sns.relplot(x = 'Instructional_research_and_public_service_FTE', 
            y = 'Total_FTE_staff',
            hue = 'Carnegie_Classification',
            style = 'Carnegie_Classification',
            data = data,
            height=6, aspect=2)
```

```python
g = sns.relplot(x = 'Instructional_research_and_public_service_FTE', 
            y = 'Total_FTE_staff',
            size = 'Average_net_pricestudents_awarded_grant_or_scholarship_aid_201718',
            hue = 'Average_net_pricestudents_awarded_grant_or_scholarship_aid_201718',
            palette="ch:r=-0.6,l=.75,d=0,h=1",
            data = data,
            sizes=(15, 200),
            height=6, aspect=2)

#box = g.ax.get_position()
#g.ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
#g.ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

g._legend.texts[0].set_text("Avg price")

g.fig.suptitle("My Title", x=0.3, y=0.98)
plt.subplots_adjust(top=0.85, left=0.1, right=0.5)

g._legend.set_bbox_to_anchor([0.17, 0.65])  # coordinates of lower left of bounding box

#plt.xlabel('x')
#plt.ylabel('y')
```

# Plotting with categorical data

- Categorical scatterplots
- Distributions of observations within categories

```python
g = sns.catplot(x='Sector_of_institution', y = 'Average_net_pricestudents_awarded_grant_or_scholarship_aid_201718',
                hue = 'Carnegie_Classification',
                kind = 'swarm',
                #kind = 'box',
                #kind = 'boxen',
                #kind = 'violin',
                #kind = 'bar',
                #kind = 'point',
                data = data, height = 7, aspect=1.5)
plt.ylabel('Avg price')
```

## If we're considering only two classes

- We can split the violin plots for better visualization

```python
g = sns.catplot(hue='Sector_of_institution', 
                y = 'Average_net_pricestudents_awarded_grant_or_scholarship_aid_201718',
                x = 'Carnegie_Classification',
                kind = 'violin',
                #kind = 'point',
                data = data, 
                split = True,
                #inner = 'stick',
                height = 7, aspect=2)

plt.ylabel('Avg price')
```

## Create a new dataframe

- And we can easily compare everybody against everybody

```python
salaries = (data.iloc[:, 149:154]/1000).drop(columns=['Average_salary_equated_to_9_months_of_fulltime_instructional_staff__instructors']).join(data.iloc[:, 13:16])
salaries.columns = ['professors', 'associate', 'assistant', 'lecturers', 'sector','carnegie','landgrant']

sns.pairplot(salaries, hue="sector");
```

# Visualizing linear relationships

- Consider confidence intervals
- Robustness whenever dealing with outliers (we need `statsmodels` installed for this to work)

```python
sns.lmplot(x="professors", y="lecturers", data=salaries,
           #ci=None, 
           #order=2,
           #robust=True,
           scatter_kws={"s": 80},
           height=6, aspect=2);
```

## The power of the facets

```python
sns.lmplot(x="professors", y="lecturers", 
           data=salaries,
           hue = 'sector',
           row = 'carnegie',
           col = 'landgrant',
           #order=2,
           #robust=True,
           scatter_kws={"s": 80},
           markers = ['^', 'o'],
           height=6, aspect=2);
```
